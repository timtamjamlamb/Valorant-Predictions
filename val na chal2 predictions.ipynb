```python
import pandas as pd
import numpy as np
```

```python
matches = pd.read_csv("valNA2022stage2matches.csv", index_col=0)
matches.head()
```

```python
matches['score'] = matches['score'].str.replace('-',' : ')
new = matches["score"].str.split(":", expand = True)
matches["GF"] = new[0]
matches["GA"] = new[1]
matches["GF"] = matches["GF"].str.replace('\xa0',' ')
matches.head()
```

```python
matches["GF"] = pd.to_numeric(matches["GF"], errors='coerce')
matches["GA"] = pd.to_numeric(matches["GA"], errors='coerce')
matches.head()
```

```python
matches['result'] = np.where(matches["GF"] > matches["GA"], "W", "L")
matches.head()
```

```python
matches.dropna(inplace = True)
matches.dtypes
```

```python
matches['opponent'] = matches['opponent'].astype(str)
matches['team'] = matches['team'].astype(str)
matches.dtypes
```

```python
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
test = matches[matches["date"] >= '2022-01-01']
```

```python
predictors = ["opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    #rolling_stats = group[cols].rolling(3, closed='left').mean()
    rolling_stats = group[cols].rolling(3).mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["GF", "GA"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-02-11']
    test = data[data["date"] >= '2022-02-11']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```
